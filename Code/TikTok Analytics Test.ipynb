# Create Helper Fucntion to Process Data

```python
# Import JSON library to laod ecport.json
import json
```

```python
# Load up data
with open('export.json','r') as f:
    data = json.load(f)
```

```python
nested_values = ['video','author', 'music','stats','authorStats','challenges','duetInfo','textExtra','stickersOnItem']
skip_values = ['challenges','duetInfo','textExtra','stickersOnItem']
```

```python
# Create blank dictionary
flattened_data = {}
# Loop thru each video
for idx, value in enumerate(data):
    flattened_data[idx] = {}
    # Loop through each value in each video
    for property_id,property_value in value.items():
        print(property_id)
        # check if nested
        if property_id in nested_values:
            # If value is to be skipped, dont add it
            if property_id in skip_values:
                pass
            else:
                # Loop through each nested property
                for nested_id, nested_value in property_value.items():
                    flattened_data[idx][property_id+'_'+nested_id] = nested_value
        else:
            # If not nested, add it back to the flattened dictionary
            flattened_data[idx][property_id] = property_value
```

```python
flattened_data[0]
```

```python
df_test = pd.DataFrame.from_dict(flattened_data, orient='index')
```
